# Score Analysis of the Q-learning agent

## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

## Analysis of the different features

```python
path_all_features = os.path.join("scores_a0.2.txt") # Using ghost and food features
path_random_actions = os.path.join("scores_random_actions.txt") # Using random actions
path_no_food_features = os.path.join("scores_no_food_features.txt") # Using only food features
path_no_ghost_features = os.path.join("scores_no_ghost_features.txt") # Using only ghost features
path_no_food_no_ghost_features = os.path.join("scores_no_food_no_ghost_features.txt") # Using neither ghost nor food features (equivalent to constant random Q-function)
```

```python
def read_data(filename):
    with open(filename, "r") as f:
        lines = f.readlines()
    
    scores = []
    for score in lines:
        value = int(score.split(' ')[1])
        scores.append(value)
    return np.array(scores)
```

```python
scores_all_features = read_data(path_all_features)
scores_random_actions = read_data(path_random_actions)
scores_no_food = read_data(path_no_food_features)
scores_no_ghost = read_data(path_no_ghost_features)
scores_no_food_no_ghost = read_data(path_no_food_no_ghost_features)
```

```python
plt.figure(figsize=(9, 6))
plt.plot(scores_all_features, label="ghost and food")
plt.plot(scores_no_food, label="only ghost")
plt.plot(scores_no_ghost, label="only food")
plt.plot(scores_no_food_no_ghost, label="no food nor ghost")
plt.title("Comparision of Q-learning agent's performance according to features used")
plt.xlabel("Number of episodes")
plt.ylabel("Score")
plt.legend()
```

```python
values = [scores_all_features.mean(), scores_no_food.mean(), scores_no_ghost.mean(), scores_no_food_no_ghost.mean(), scores_random_actions.mean()]
names = ["ghost and food", "only ghost", "only food", "no food nor ghost", "random actions"]

plt.figure(figsize=(9, 6))
plt.bar(names, values)
plt.title("Comparision of average score according to features used")
plt.ylabel("Average score")
```
